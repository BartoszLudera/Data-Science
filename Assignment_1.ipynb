## Exercise 3
a) Create boxplot of apartment's grade versus it's price. Compare with plot 2b).  
b) Create mosaic plot of any two categorical variables.

```python
df_box = df[["grade", "price"]]

# Create scatter plot
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)  # First subplot
plt.scatter(df_box["grade"], df_box["price"])
plt.title("Scatter Plot: Grade vs Price")
plt.xlabel("Grade")
plt.ylabel("Price")

# Create box plot
df_box.boxplot(column="price", by="grade")
plt.title("Box Plot: Price by Grade")
plt.suptitle("")  # Suppress the default title to avoid overlap
plt.xlabel("Grade")
plt.ylabel("Price")

# Show plots
plt.show()
```

```python
#b)
from statsmodels.graphics.mosaicplot import mosaic

#Your code:

mosaic(df, ['grade', 'condition'], title='Grade to condition')


```
